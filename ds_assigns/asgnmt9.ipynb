```python
# Transform the original dataset into the principal components
from sklearn.decomposition import PCA
pca = PCA()
PC = pca.fit_transform(df)
PC
```

```python
# Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
explained_variance = pca.explained_variance_ratio_

# Scree plot
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, edgecolor="black")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()

# Determine the cumulative explained variance
import numpy as np
cumulative_explained_variance = np.cumsum(explained_variance)

# Print the cumulative explained variance
print("Cumulative Explained Variance:")
for i, variance in enumerate(cumulative_explained_variance):
  print(f"  {i+1} components: {variance:.3f}")

# Choose the optimal number of principal components based on the scree plot and cumulative explained variance
optimal_components = 2  # Replace with the desired number of components
```

```python
# Transform the original dataset into the principal components
PC_new = PC[:, :optimal_components]

# Print the transformed dataset
print(PC_new)
```

```python
#  Clustering with Original Data:
#	Apply a clustering algorithm (e.g., K-means) to the original dataset
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example dataset creation
data = {
    'Alcohol': [14.23, 13.2, 13.16, 14.37, 13.24, 14.2, 14.39, 14.06, 14.83, 13.86],
    'Color': [5.64, 4.38, 5.68, 7.8, 4.32, 6.75, 6.8, 5.0, 5.7, 4.1],
    'Proline': [1065, 1050, 1185, 1480, 735, 1450, 1290, 1295, 1045, 845]
}
df = pd.DataFrame(data)

# Fit the KMeans model
num_clusters = 3  # Specify the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(df)

# Define the colors for the clusters
colors = ['red', 'green', 'blue']

# 2D scatter plot of the clusters
for i in range(num_clusters):
    plt.scatter(df[kmeans.labels_ == i]["Alcohol"], df[kmeans.labels_ == i]["Color"], c=colors[i])

plt.xlabel("Alcohol")
plt.ylabel("Color")
plt.title("Clustered Data")
plt.show()

# Plot the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='yellow', label='Centroids')
plt.xlabel("Alcohol")
plt.ylabel("Color")
plt.title("Clustered Data with Centroids")
plt.legend()
plt.show()

# 3D scatter plot of the clusters
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i in range(num_clusters):
    ax.scatter(df[kmeans.labels_ == i]["Alcohol"], df[kmeans.labels_ == i]["Color"], df[kmeans.labels_ == i]["Proline"], c=colors[i])

ax.set_xlabel("Alcohol")
ax.set_ylabel("Color")
ax.set_zlabel("Proline")
ax.set_title("Clustered Data in 3D")
plt.show()
```

```python
# Visualize the clustering results using appropriate plots.
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example dataset creation
data = {
    'Alcohol': [14.23, 13.2, 13.16, 14.37, 13.24, 14.2, 14.39, 14.06, 14.83, 13.86],
    'Color': [5.64, 4.38, 5.68, 7.8, 4.32, 6.75, 6.8, 5.0, 5.7, 4.1],
    'Proline': [1065, 1050, 1185, 1480, 735, 1450, 1290, 1295, 1045, 845]
}
df = pd.DataFrame(data)
# Fit the KMeans model
num_clusters = 3  # Specify the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(df)
# Define the colors for the clusters
colors = ['red', 'green', 'blue']

# 2D scatter plot of the clusters
for i in range(num_clusters):
    plt.scatter(df[kmeans.labels_ == i]["Alcohol"], df[kmeans.labels_ == i]["Color"], c=colors[i], label=f'Cluster {i+1}')

plt.xlabel("Alcohol")
plt.ylabel("Color")
plt.title("Clustered Data")
plt.legend()
plt.show()
# Plot the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='yellow', label='Centroids', marker='X', s=100)
plt.xlabel("Alcohol")
plt.ylabel("Color")
plt.title("Clustered Data with Centroids")
plt.legend()
plt.show()
# 3D scatter plot of the clusters
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i in range(num_clusters):
    ax.scatter(df[kmeans.labels_ == i]["Alcohol"], df[kmeans.labels_ == i]["Color"], df[kmeans.labels_ == i]["Proline"], c=colors[i], label=f'Cluster {i+1}')

ax.set_xlabel("Alcohol")
ax.set_ylabel("Color")
ax.set_zlabel("Proline")
ax.set_title("Clustered Data in 3D")
ax.legend()
plt.show()
```

```python
# Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Calculate the silhouette score
silhouette_avg = silhouette_score(df, kmeans.labels_)
print("Silhouette Score:", silhouette_avg)

# Calculate the Davies-Bouldin index
davies_bouldin = davies_bouldin_score(df, kmeans.labels_)
print("Davies-Bouldin Index:", davies_bouldin)
```

```python
kmeans_pca = KMeans(n_clusters=num_clusters)

# Fit the model to the PCA-transformed dataset
kmeans_pca.fit(PC_new)

# Print the cluster labels for each data point
print(kmeans_pca.labels_)

# Visualize the clusters
plt.scatter(PC_new[:, 0], PC_new[:, 1], c=kmeans_pca.labels_)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Clustered Data (PCA)")
plt.show()

# Calculate the silhouette score
silhouette_avg_pca = silhouette_score(PC_new, kmeans_pca.labels_)
print("Silhouette Score (PCA):", silhouette_avg_pca)

# Calculate the Davies-Bouldin index
davies_bouldin_pca = davies_bouldin_score(PC_new, kmeans_pca.labels_)
print("Davies-Bouldin Index (PCA):", davies_bouldin_pca)
```

```python
# Plot the clusters using different colors
colors = ['red', 'green', 'blue']

for i in range(num_clusters):
  plt.scatter(PC_new[kmeans_pca.labels_ == i, 0], PC_new[kmeans_pca.labels_ == i, 1], c=colors[i])

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Clustered Data (PCA)")
plt.show()

# Plot the centroids of the clusters
plt.scatter(kmeans_pca.cluster_centers_[:, 0], kmeans_pca.cluster_centers_[:, 1], c='yellow', label='Centroids')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Clustered Data with Centroids (PCA)")
plt.legend()
plt.show()
```

```python
# Compare the clustering results

# Compare the silhouette scores
print("Silhouette Score (Original Data):", silhouette_avg)
print("Silhouette Score (PCA):", silhouette_avg_pca)

# Compare the Davies-Bouldin indices
print("Davies-Bouldin Index (Original Data):", davies_bouldin)
print("Davies-Bouldin Index (PCA):", davies_bouldin_pca)

# Visualize the clusters for both datasets
plt.figure(figsize=(15, 5))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(df["Alcohol"], df["Color"], c=kmeans.labels_)
plt.xlabel("Alcohol")
plt.ylabel("Color")
plt.title("Original Data")

# PCA-transformed data
plt.subplot(1, 2, 2)
plt.scatter(PC_new[:, 0], PC_new[:, 1], c=kmeans_pca.labels_)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA-Transformed Data")
plt.show()
```
