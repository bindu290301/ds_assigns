```python
##	Create bar charts to visualize the frequency or count of each category
import matplotlib.pyplot as plt
# Bar graph for date
df.groupby('Date')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales by Date")
plt.show()

# Bar graph for date
df.groupby('Day')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Day")
plt.ylabel("sales")
plt.title("Sales by Day")


# Bar graph for sku
df.groupby('SKU')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("SKU")
plt.ylabel("Sales")
plt.title("Sales by SKU")
plt.show()

# Bar graph for city
df.groupby('City')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Sales")
plt.title("Sales by City")
plt.show()

# Bar graph for bu
df.groupby('BU')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("BU")
plt.ylabel("Sales")
plt.title("Sales by BU")
plt.show()

# Bar graph for brand
df.groupby('Brand')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Brand")
plt.ylabel("Sales")
plt.title("Sales by Brand")
plt.show()

# Bar graph for model
df.groupby('Model')['Total Sales Value'].sum().plot(kind='bar')
plt.xlabel("Model")
plt.ylabel("Sales")
plt.title("Sales by Model")
plt.show()

```

```python
## Standardization of Numerical Variables
==>The concept of standardization (z-score normalization)
    Standardization, also known as z-score normalization, is a technique used to rescale features of data to have the properties
    of a standard normal distribution with a mean of 0 and a standard deviation of 1.
    formula: z=x-mu/sigma
```

```python
## Conversion of Categorical Data into Dummy Variables
The need for converting categorical data into dummy variables (one-hot encoding)
. Machine Learning Algorithms Require Numerical Input
. Avoiding Ordinal Relationships
. Ensuring Equal Distance
. Handling Multiple Categories                                                               
```

```python
## one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
import pandas as pd
categorical_columns = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_columns, dtype=int)
df.head()
```
